```python
vgg = VGG19(input_shape = (224,224,3), include_top=False, weights='imagenet')
```

```python
model = Sequential()
```

```python
for layers in vgg.layers:
    layers.trainable=False
```

```python
for layers in vgg.layers:
    model.add(layers)
```

```python
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```
